```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/streaming_data.csv')
```

```python
dados.head()
```

```python
dados.loc[:, ['Time_on_platform', 'Num_streaming_services', 'Churned', 'Avg_rating', 'Devices_connected']] = dados.loc[:,['Time_on_platform', 'Num_streaming_services', 'Churned', 'Avg_rating', 'Devices_connected']].fillna(0)
```

```python
dados.describe()
```

```python
dados.info()
```

```python
dados.isnull().sum()
```

```python
dados = dados.dropna()
```

```python
dados.head()
```

```python
dados['Churned'] = dados['Churned'].replace({0: 'No', 1: 'Yes'})
```

```python
dados.head()
```

```python
dados.dtypes
```

```python
dados[['Age', 'Time_on_platform', 'Devices_connected', 'Num_streaming_services', 'Avg_rating']] = dados[['Age', 'Time_on_platform', 'Devices_connected', 'Num_streaming_services', 'Avg_rating']].astype(int)
```

```python
dados_plv = dados[['Gender', 'Subscription_type',]]
dados_num = dados[['Age',	'Time_on_platform',	'Devices_connected',	'Num_streaming_services',	'Num_active_profiles',	'Avg_rating']]

dados_total = pd.concat([dados_plv, dados_num], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

le_gender = LabelEncoder()
le_subscription = LabelEncoder()

dados_plv['Gender'] = le_gender.fit_transform(dados_plv['Gender'])
dados_plv['Subscription_type'] = le_subscription.fit_transform(dados_plv['Subscription_type'])

dados_plv = pd.get_dummies(dados_plv, columns=['Gender', 'Subscription_type'])

dados_total = pd.concat([dados_plv, dados_num], axis=1)
```

```python
X = dados_total.copy()
y = dados[['Churned']]


le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LogisticRegression(class_weight='balanced')
lr = model.fit(X_train, y_train)
```

```python
y_pred_proba = model.predict_proba(X_test)[:, 1]

threshold = 0.3
y_pred = (y_pred_proba >= threshold).astype(int)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import confusion_matrix

y_pred_prob = model.predict_proba(X_test)[:, 1]
y_pred = (y_pred_prob > 0.5).astype(int)

conf_matrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Classe 0', 'Classe 1'], yticklabels=['Classe 0', 'Classe 1'],
            linewidths=1, linecolor='black')

plt.title('Matriz de Confuso', fontsize=16)
plt.xlabel('Classe Predita', fontsize=14)
plt.ylabel('Classe Real', fontsize=14)

plt.show()
```

```python
from sklearn.metrics import (
    accuracy_score,
    balanced_accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
```

```python
print(f"Acurcia (Treino): {accuracy_score(y_train, lr.predict(X_train))}")
print(f"Acurcia (Teste): {accuracy_score(y_test, lr.predict(X_test))}")
print("====================================")
print(f"Acurcia Balanceada (Treino): {balanced_accuracy_score(y_train, lr.predict(X_train))}")
print(f"Acurcia Balanceada (Teste): {balanced_accuracy_score(y_test, lr.predict(X_test))}")
print("====================================")
print(f"Precision (Treino): {precision_score(y_train, lr.predict(X_train))}")
print(f"Precision (Teste): {precision_score(y_test, lr.predict(X_test))}")
print("====================================")
print(f"Recall (Treino): {recall_score(y_train, lr.predict(X_train))}")
print(f"Recall (Teste): {recall_score(y_test, lr.predict(X_test))}")
print("====================================")
print(f"F1-Score (Treino): {f1_score(y_train, lr.predict(X_train))}")
print(f"F1-Score (Teste): {f1_score(y_test, lr.predict(X_test))}")
print("====================================")
print(f"ROCAUC (Treino): {roc_auc_score(y_train, lr.predict_proba(X_train)[:,1])}")
print(f"ROCAUC (Teste): {roc_auc_score(y_test, lr.predict_proba(X_test)[:,1])}")
```
